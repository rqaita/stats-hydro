```python
# Line
serie_anual.plot(x=0, y=1, xlabel='Ano', ylabel='Vazão (m$^3$/s)', legend=None)
```

```python
# Bar
ax = serie_anual.plot.bar(x=0, y=1, legend=None, width=0.9)
ax.set_xlabel('Ano')
ax.set_xticks(np.arange(0, 70, 10))
ax.set_xticklabels(np.arange(1938, 2000, 10), rotation=0)
ax.set_ylabel('Vazão (m$^3$/s)')
```

```python
# Histogram
N = serie_anual.shape[0]
NC = round(1 + 3.3 * np.log10(N))  # Sturges
counts, bins = np.histogram(serie_anual[1], bins=NC)

ax = serie_anual.plot.hist(y=1, bins=NC, legend=None)
ax.set_xlabel('Vazão (m$^3$/s)')
ax.set_xticks(bins)
ax.set_ylabel('Frequência Absoluta')
```

```python
# Frequency polygon
counts, bins = np.histogram(serie_anual[1], bins=NC)
bins = (bins[:-1] + bins[1:]) / 2

fig, ax = plt.subplots()
ax.plot(bins, counts)
ax.set_xlabel('Vazão (m$^3$/s)')
ax.set_ylabel('Frequência Absoluta')
```

```python
# Flow duration curve
vazao_decr = serie_anual[1].sort_values(ascending=False)
N = vazao_decr.shape[0]
n = np.arange(N) + 1
n_N = n / N

fig, ax = plt.subplots()
ax.plot(n_N * 100, vazao_decr)
ax.set_xlabel('Permanência (%)')
ax.set_ylabel('Vazão (m$^3$/s)')
```
